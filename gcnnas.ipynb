# Gridsearch

```python
kwargs = {
    'dataset_name':'Cora',
    'epochs':500,
    'config':config,
    'merge_acts':True,
    'prune':False,
}

results = {}
lrs = [0.1 , 0.01, 0.001, 0.0005]
wds = [1e-3, 5e-4, 1e-5, 5e-6]
opts = ['adam', 'sgd', 'rmsprop']
dropouts = [0.0, 0.25, 0.5, 0.75]
widths = [8, 16, 32, 64, 128, 256]
print(len(lrs)*len(wds)*len(opts)*len(dropouts)*len(widths))

for lr in lrs:
    for wd in wds:
        for opt in opts:
            for dropout in dropouts:
                for width in widths:
                    kwargs['lr'] = lr
                    kwargs['wd'] = wd
                    kwargs['opt'] = opt
                    kwargs['dropout'] = dropout
                    kwargs['width'] = width
                    kwargs['init'] = lr if opt!='rmsprop' else 10*lr
                    
                    print("==",lr,wd,opt,dropout,width,"==")
                    manager = GNNTrainer(kwargs)
                    res = manager.train(verbose=False)
                    print(max(res['test_acc']))
                    results[(lr,wd,opt,dropout)] = res
```

```python
max([(k,max(v['test_acc'])) for k,v in results.items()], key=lambda x: x[1])
```
